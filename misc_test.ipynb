```python
import os
import zipfile
import requests

url = "https://data.deepai.org/conll2003.zip"
save_dir = "data/conll2003_raw"
os.makedirs(save_dir, exist_ok=True)

zip_path = os.path.join(save_dir, "conll2003.zip")

# 下载 zip
with requests.get(url, stream=True, timeout=60) as r:
    r.raise_for_status()
    with open(zip_path, "wb") as f:
        for chunk in r.iter_content(chunk_size=8192):
            if chunk:
                f.write(chunk)

# 解压
with zipfile.ZipFile(zip_path, "r") as z:
    z.extractall(save_dir)

print("下载并解压完成")
print("zip 文件：", zip_path)
print("解压目录：", save_dir)
print("目录内容：", os.listdir(save_dir))
```

```python
from utils.bilstm_crf.data_utils import read_conll_4

train_file_path = 'data/conll2003_raw/data/train.txt'
test_file_path = 'data/conll2003_raw/data/test.txt'
valid_file_path = 'data/conll2003_raw/data/valid.txt'

train_save_path = 'data/conll2003_processed/train.txt'
valid_save_path = 'data/conll2003_processed/valid.txt'
test_save_path = 'data/conll2003_processed/test.txt'

def save_conll2(sentences, ner_tags, save_path):
    with open(save_path, "w", encoding="utf-8") as f:
        for words, tags in zip(sentences, ner_tags):
            for w, t in zip(words, tags):
                f.write(f"{w} {t}\n")
            f.write("\n")

def func(_file_path, _save_path):
    sentence, tag_list = read_conll_4(_file_path)
    save_conll2(sentence, tag_list, _save_path)
    return

func(train_file_path, train_save_path)
func(test_file_path, test_save_path)
func(valid_file_path, valid_save_path)
```
